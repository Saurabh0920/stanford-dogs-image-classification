**Purpose:** Train a small CNN from scratch on the Stanford Dogs dataset without augmentation.

**Notebook structure:**

- Data generator (no augmentation)
- Model definition
- Training loop
- Evaluation & results

```
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print('TensorFlow version:', tf.__version__)
```

```
# --- Data paths (edit to your local paths) ---
DATA_DIR = '/path/to/stanford_dogs'
TRAIN_DIR = os.path.join(DATA_DIR, 'train')
VAL_DIR = os.path.join(DATA_DIR, 'val')

# Use tf.keras.preprocessing.image_dataset_from_directory or image generators
```

```
# --- Model (small CNN) ---
def build_basic_cnn(input_shape=(224,224,3), n_classes=120):
    model = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(64,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(n_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

model = build_basic_cnn()
model.summary()
```
